# FEATURE DROPPING FOR PERINUCLEAR FILE USING HEATMAPS 

```python
df_perinuclear = pd.read_pickle('/Users/apple/Desktop/DATA_590/cleaned_data/PAM194_Keratino_CytoPanel_1/pam194ObjPerinuclear.pkl')
```

```python
df_perinuclear.head()
```

### With reference from the correlation heatmaps dashboard
1. Drop 3 out of 4 texture_Contrast for same type (i.e DNA, CorrWGA, CorrMito etc)
2. Drop 3 out of 4 Texture_AngularSecondMoment for same type (i.e DNA, CorrWGA, CorrMito etc)

```python
def drop_columns(data):
    data.drop(list(data.filter(regex = 'FileName_')), axis=1, inplace=True) # Dropping all of the columns starting with 'FileName_'
    data.drop(list(data.filter(regex = 'PathName_')), axis=1, inplace=True) # Dropping all of the columns starting with 'PathName_'
    # data.drop(['Metadata_Date', 'Metadata_FileLocation', 'Metadata_Frame', 'Metadata_Run', 'Metadata_Series'], axis=1, inplace=True)
    
```

```python
drop_columns(df_perinuclear)
df_perinuclear.head()
```

##Keep one from texture contrast and one from texture angular second moment drop all others 

```python
# Select columns starting with "columns_of_interest_for_heatmap"
column_prefix = 'Texture_Contrast'
all_columns = df_perinuclear.filter(regex=f'^{column_prefix}_', axis=1).columns.tolist()
```

```python
all_columns
```

```python
columns_to_delete = []
for i in range(len(all_columns)):
    if i % 4 != 0:
        columns_to_delete.append(all_columns[i])
    
```

```python
columns_to_delete
```

```python
df_perinuclear.drop(columns_to_delete, axis = 1, inplace = True)
```

```python
df_perinuclear.shape
```

```python
# Select columns starting with "columns_of_interest_for_heatmap"
column_prefix = 'Texture_AngularSecondMoment'
all_columns = df_perinuclear.filter(regex=f'^{column_prefix}_', axis=1).columns.tolist()
```

```python
all_columns
```

No column to drop for **Texture_AngularSecondMoment** as all were dropped in PCA 

```python
df_perinuclear.to_pickle('/Users/apple/Desktop/DATA_590/cleaned_data/PAM194_Keratino_CytoPanel_1/pam194ObjPerinuclear_final.pkl')
```
